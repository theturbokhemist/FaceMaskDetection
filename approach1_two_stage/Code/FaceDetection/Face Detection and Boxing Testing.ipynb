## Pre Recs

These images have both masks and no masks. Some images have 1 person per image and some have 2+.

```python
# some images from the dataset are stored in the test image directory
img_dir = './test_imgs/'
img_names = []
for i in range(1, 13):
    name = 'face' + str(i)
    full_name = img_dir + name
    if i in [1, 2, 3, 4]:
        full_name = full_name + '.jpg'
    else:
        full_name = full_name + '.png'
    img_names.append(full_name)
```
